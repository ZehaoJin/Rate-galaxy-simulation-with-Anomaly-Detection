```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
feature=pd.read_csv('feature_archive/SDSS.csv')
m=pd.read_csv('t_SDSS.csv')
```

```python
feature=feature.sort_values('galaxy')
```

```python
m=m.sort_values('name')
```

```python
feature
```

```python
m
```

```python
feature['m']=m['m']
```

```python
feature
```

```python
x=feature.iloc[:,2:-1]
#x=np.log10(np.abs(feature.iloc[:,2:]))
y=np.log10(feature.iloc[:,-1])
#y=m.iloc[:,1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=40)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

# Define the neural network model
model = Sequential()
model.add(Dense(64, input_dim=128, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

# Compile the model with mean squared error loss and Adam optimizer
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model on the training data with a progress bar
history = model.fit(x_train, y_train, epochs=50,batch_size=128,validation_data=(x_test, y_test))
```

```python
y_pred=model.predict(x_test)
y_pred_train=model.predict(x_train)
```

```python
plt.scatter(y_train,y_pred_train,label='training set',alpha=0.7)
plt.scatter(y_test,y_pred,label='test set',alpha=0.5)
line=np.linspace(np.array([y_test.min(),y_pred.min()]).min(),np.array([y_test.max(),y_pred.max()]).max())
plt.plot(line,line,ls='--')
plt.xlabel('truth')
plt.ylabel('prediction')
#plt.xscale('log')
#plt.yscale('log')
plt.legend()
```

```python
# Define the neural network model
model = keras.models.Sequential([
    keras.layers.Dense(1, activation="linear", input_shape=(128,), kernel_regularizer=keras.regularizers.l1(0.01))
])

# Compile the model with mean squared error loss and Adam optimizer
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model on the training data with a progress bar
history = model.fit(x_train, y_train, epochs=100,batch_size=128,validation_data=(x_test, y_test))
```

```python
y_pred=model.predict(x_test)
y_pred_train=model.predict(x_train)
```

```python
plt.scatter(y_train,y_pred_train,label='training set',alpha=0.7)
plt.scatter(y_test,y_pred,label='test set',alpha=0.5)
line=np.linspace(np.array([y_test.min(),y_pred.min()]).min(),np.array([y_test.max(),y_pred.max()]).max())
plt.plot(line,line,ls='--')
plt.xlabel('truth')
plt.ylabel('prediction')
#plt.xscale('log')
#plt.yscale('log')
plt.legend()
```

```python
# Define the neural network model with L1 regularization
model = keras.models.Sequential([
    keras.layers.Dense(128, activation="relu", input_shape=(128,), kernel_regularizer=keras.regularizers.l1(0.01)),
    keras.layers.Dense(64, activation="relu", kernel_regularizer=keras.regularizers.l1(0.01)),
    keras.layers.Dense(32, activation="relu", kernel_regularizer=keras.regularizers.l1(0.01)),
    keras.layers.Dense(1, activation="linear", kernel_regularizer=keras.regularizers.l1(0.01))
])

# Compile the model with mean squared error loss and Adam optimizer
model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model on the training data with a progress bar
history = model.fit(x_train, y_train, epochs=100,batch_size=128,validation_data=(x_test, y_test))
```

```python
y_pred=model.predict(x_test)
y_pred_train=model.predict(x_train)
```

```python
plt.scatter(y_train,y_pred_train,label='training set',alpha=0.7)
plt.scatter(y_test,y_pred,label='test set',alpha=0.5)
line=np.linspace(np.array([y_test.min(),y_pred.min()]).min(),np.array([y_test.max(),y_pred.max()]).max())
plt.plot(line,line,ls='--')
plt.xlabel('truth')
plt.ylabel('prediction')
#plt.xscale('log')
#plt.yscale('log')
plt.legend()
```
